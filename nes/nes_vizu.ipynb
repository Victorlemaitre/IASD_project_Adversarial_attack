```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import re
import pandas as pd
```

```python
csv_path = "assets/sweep_L_sigma0.001_ns50_eps0.1_delta0.02_Nimages200.csv"
df = pd.read_csv(csv_path)

plt.figure(figsize=(7,5))
plt.plot(df["L"], df["adv_acc"] * 100, marker="o")

plt.xlabel("Query budget L ()")
plt.ylabel("NES+PGD accuracy (%)")
plt.title("Attacked Accuracy vs Query Budget ($N_{images}=200$)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
csv_path = "assets/sweep_sigma_L500_ns50_eps0.1_delta0.02_Nimages200.csv"
df_sweep_NES = pd.read_csv(csv_path)
csv_path_2 = "../pgd/assets/pgd_accuracy_delta_0.02_iters_5.csv"
df_sigma_PGD = pd.read_csv(csv_path_2)

plt.figure(figsize=(7,5))
plt.plot(
    df_sweep_NES["epsilon"],
    df_sweep_NES["adv_acc"] * 100,
    marker="o",
    label="Black-box NES+PGD"
)
plt.plot(
    df_sigma_PGD["epsilon"],
    df_sigma_PGD["accuracy"],
    marker="o",
    label="White-box PGD"
)
plt.xlabel("Epsilon")
plt.ylabel("Accuracy (%)")
plt.title("Attacked accuracy vs epsilon on CIFAR-10 (delta=0.02, iters=5)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
